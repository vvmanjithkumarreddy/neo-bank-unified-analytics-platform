# Create Catalog and Schema
1. Create Schema landing, bronze, silver, gold
2. Create Volume files in landing 

```sql
CREATE SCHEMA IF NOT EXISTS neo_bank.landing;

CREATE SCHEMA IF NOT EXISTS neo_bank.bronze;

CREATE SCHEMA IF NOT EXISTS neo_bank.silver;

CREATE SCHEMA IF NOT EXISTS neo_bank.gold;
```

```sql
CREATE VOLUME neo_bank.landing.files;
```
