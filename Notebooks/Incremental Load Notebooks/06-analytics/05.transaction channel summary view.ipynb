# Transaction Channel Summary View
Create Transaction Channel Summary View with the columns
- txn_date
- gateway_name
- device_type
- total_transactions
- successful_transactions
- failed_transactions
- avg_processing_time_ms

```sql
CREATE OR REPLACE VIEW neo_bank.gold.vw_transaction_channel_summary AS 

select
    full_date as txn_date,
    gateway_name,
    device_type,
    count(txn_id) as total_transactions,
    sum(CASE WHEN gateway_status=='SUCCESS' THEN 1 ELSE 0 END) as successful_transactions,
    sum(CASE WHEN gateway_status=='FAILED' THEN 1 ELSE 0 END) as failed_transactions,
    avg(processing_time_ms) as avg_processing_time_ms
from neo_bank.gold.fact_payment_gateway_logs
inner join neo_bank.gold.dim_date
on neo_bank.gold.fact_payment_gateway_logs.processed_date_key==neo_bank.gold.dim_date.date_key
group by txn_date,gateway_name,device_type
order by txn_date,gateway_name,device_type
```
