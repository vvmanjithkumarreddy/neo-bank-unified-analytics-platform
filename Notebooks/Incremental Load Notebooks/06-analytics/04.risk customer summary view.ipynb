# Risk Customer Summary View
Create risk customer summary view with the columns
- risk_grade
- total_customers
- avg_credit_score
- total_external_loans
- total_overdue_amount

```sql
CREATE OR REPLACE VIEW neo_bank.gold.vw_risk_customer_summary AS 

SELECT 
    risk_grade, 
    count(customer_sk) as total_customers, 
    avg(credit_score) as avg_credit_score,
    sum(external_active_loans) as total_external_loans,
    sum(external_overdue_amount) as total_overdue_amount
from neo_bank.gold.fact_credit_bureau_reports
group by risk_grade 
ORDER BY 
    CASE risk_grade
        WHEN 'LOW' THEN 3
        WHEN 'MEDIUM' THEN 2
        WHEN 'HIGH' THEN 1
    END
```
